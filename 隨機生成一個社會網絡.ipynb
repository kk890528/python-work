# 畫出一個社會網絡
### 用隨機抽的方式，畫出一個有三個中心點的社會網絡

```python
people=['A','B','C','D','E','F','G','H','I','J','K','M','N','O','P','Q','R','S','T']
```

```python
np.random.choice(people,10,replace=False)
```

```python
df=pd.DataFrame({'個體1':list('A'*11+'B'*7+'C'*9)})
```

```python
pop2=np.append(np.random.choice(people,11,replace=False),np.random.choice(people,7,replace=False))
pop2=np.append(pop2,np.random.choice(people,9,replace=False))
```

```python
df['個體B']=pop2
```

```python
df['好感度']=np.random.randint(0,16,len(df))
```

```python
df.head()
```

```python
df.iloc[11,0]
```

```python
G=nx.Graph()
for i in range(len(df)):
    G.add_edge(df.iloc[i,0],df.iloc[i,1], weight = df.iloc[i,2],color='green')
```

```python
plt.figure(figsize=(16,7))

pos=nx.spring_layout(G) 

nx.draw_networkx_nodes(G,pos,node_size=1000, alpha=0.5)

nx.draw_networkx_edges(G,pos,alpha=0.5, edge_color ='b')

nx.draw_networkx_labels(G,pos,font_size=13,font_color='w')

edgewidth = [d['weight'] for (u,v,d) in G.edges(data=True)]
nx.draw_networkx_edges(G,pos,alpha=0.5, edge_color ='b', width=edgewidth)
```
