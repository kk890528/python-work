# 中國信用制度
## 為了成為1984的老大哥，中國將要開始實施信用制度，初始點數為1000

```python
class credit:
    def __init__(self, name):
        try:
            self.name = name
            self.credit_value = 1000
            self.record=pd.DataFrame(columns=["原因","增減"],)
            self.behave={"詆毀席維尼":-2000,"加入共產黨青年軍":200,"超速":-50,"縹妓":-500,"捐錢給黨":1000,"不愛國":-1000,
                        "舉發不愛國人士":300,"舉發台獨":500,"向國旗敬禮":50,"買太多遊戲":-100,"當任志工":50}
        except:
            print("查無此人")
    def behavior(self,reason):
        try:
            if reason in ["支持台獨","支持港獨","支持藏獨","支持新疆","宣傳六四"]:
                self.credit_value = 0
                print("我們懷念你")
                del(self.record)
                del(self.credit_value)
                del(self.name)
                del(self.behave)
            else:
                    self.credit_value+=self.behave[reason]
                    self.record.loc[self.record.shape[0]+1]=[reason,f"{self.behave[reason]}"]
                    if self.credit_value<=0:
                        self.credit_value=0
                        print("親，你破產了")
                    else:
                        print(f"信用點數{self.behave[reason]}，目前剩餘{self.credit_value}")
        except:
            print("查無此人")
    def __repr__(self):
            try:
                value=f"姓名:  {self.name}\n\n目前您的信用點數為:  {self.credit_value}點"
                return(value)
            except AttributeError:
                return("查無此人")
```

```python
c=credit("呂修齊")
```
