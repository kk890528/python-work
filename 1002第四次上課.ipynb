```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from numpy.random import randint
from numpy.random import choice
```

```python
for i in range(50):
    print(randint(1,10),end=",")
```

```python
words='''
我
我的
妳
妳的
心
溫柔
日子
雨
風
天空
雲
等待
哭泣
戀愛
相遇
分離
忘記
心醉
驀然
吹過
思念
靈魂
停止'""
```

```python
phrase=words.split('\n')
```

```python
phrase=phrase[1:]
```

```python
np.random.choice(phrase,5)
```

```python
egg=list()
n1=np.random.randint(3,6)
for i in range(n1):
    n2=np.random.randint(5,8)
    spam=np.random.choice(phrase,n2)
    egg.append(spam)
```

```python
for i in range(len(egg)):
    print(' '.join(egg[i]))
```

```python
ii=np.random.randint(5,8)
egg=np.random.choice(phrase,ii)
' '.join(egg)
```

```python
n=randint(3,6)
for i in range(n):
    k=randint(5,8)
    egg=np.random.choice(phrase,k)
    ham=" ".join(egg)
    print(ham)
```
