# 作業1：寫一個程式，當你查詢單字的意思時，可以幫你翻譯並記錄你查過的單字

```python
class voc:
    def __init__(self,name):
        self.name= name
        self.record=pd.DataFrame(columns=["英文","中文"])#紀錄
        self.record=self.record.rename_axis('查詢紀錄',axis=0)
    def translate(self,word): #翻譯
        translator = Translator()
        tran=translator.translate(word,src='en' ,dest = "zh-TW")
        self.record.loc[self.record.shape[0]+1]=[word,f"{tran.text}"]
        print('英文：',word,'\n'+'中文：',tran.text)
    def test_book(self): #查詢曾經錯過的單字
        test=self.record.groupby(["英文",'中文']).英文.count()
        print(test)
    def __repr__(self):
        return (self.name)
```

```python
v= voc('呂修齊的單字本')
```
