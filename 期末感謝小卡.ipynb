```python
import turtle as tu
from random import *
import time
```

# 用turtle期末感謝小卡
### 之前曾經也用一樣的方式寫給朋友生日的卡片，可以避免收到卡片覺得佔位的丟掉用很尷尬的情況發生。我把內容改成期末給老師跟助教的感謝小考，也希望避免老師助教如果收到實質的會覺得困擾的狀況XDD

```python
tu.showturtle()
tu.setup(800, 600, 0, 0)
tu.pensize(5)
tu.bgcolor("black") 
tu.pencolor('white')
tu.penup()
tu.goto(-300,200)
time.sleep(2)
tu.write("老師、助教，",True,'Center',('Arial',24,'bold'))
time.sleep(2)
tu.goto(-150,150)
tu.write("首先感謝這兩個學期的教導，",True,'Center',('Arial',24,'bold'))
time.sleep(2)
tu.goto(-100,100)
tu.write("也謝謝常常課餘時間讓我問問題打擾你們。",True,'Center',('Arial',24,'bold'))
time.sleep(2)
tu.goto(-150,50)
tu.write("由於本系跟程式的直接相關性不大",True,'Center',('Arial',24,'bold'))
tu.goto(0,0)
time.sleep(2)
tu.write("這算是第一次接觸這類程式(統計軟體不算的話)，",True,'Center',('Arial',24,'bold'))
tu.goto(-50,-50)
time.sleep(2)
tu.write("但這門課讓我萌生了對於人工智能的興趣，",True,'Center',('Arial',24,'bold'))
tu.goto(-100,-100)
time.sleep(2)
tu.write("並開始思考人工智能如何與主修做結合，",True,'Center',('Arial',24,'bold'))
tu.goto(-125,-150)
time.sleep(2)
tu.write("讓我獲益良多",True,'Center',('Arial',24,'bold'))
tu.goto(0,-200)
time.sleep(2)
tu.write("呂修齊",True,'Center',('Arial',24,'bold'))
tu.speed(1)
tu.goto(35,-165)
tu.write("敬上",True,'Center',('Arial',12,'bold'))
time.sleep(5)
tu.clear()
time.sleep(1)
def ground():
    tu.hideturtle()
    
    for i in range(400):
        tu.pensize(randint(5,10))
        x = randint(-400,350)
        y = randint(-280,-1)
        r = -y/280
        g = -y/280
        b = -y/280
        tu.pencolor((r, g, b))
        tu.penup()
        tu.goto(x,y)
        tu.pendown()
        tu.forward(randint(40,100))

def snow():    
    tu.hideturtle()
    tu.pensize(2)
    #100朵花瓣
    for i in range(100):
        r = random()
        g = random()
        b = random()
        tu.pencolor(r, g, b)
        tu.penup()
        tu.setx(randint(-350,350))
        tu.sety(randint(1,270))
        tu.pendown()
        #dens表示花瓣数
        dens = randint(8,12)
        snowsize = randint(10,14)
        for j in range(dens):
            tu.forward(snowsize)
            tu.backward(snowsize)
            tu.right(360/dens)


def main():
    tu.setup(800, 600, 0, 0)
    tu.tracer(False)
    tu.bgcolor("black")    
    snow()
    ground()
    tu.tracer(True)


if __name__ == "__main__":
    main()

time.sleep(2)
tu.penup()
tu.pensize(5)
tu.sety(0)
tu.setx(0)
tu.color("white")
tu.write("謝謝老師和助教",True,'Center',('Arial',48,'bold'))


tu.goto(250,-180)
tu.setheading(10)

def curvemove():
    for i in range(200):
        tu.right(1)
        tu.forward(1)
tu.color('maroon','red')        
tu.speed(1000)
tu.penup()
tu.begin_fill()
tu.left(140)
tu.forward(111.65)
curvemove()
tu.left(120)
curvemove()
tu.forward(111.65)
tu.end_fill()
tu.done()
```
