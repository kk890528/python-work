```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import sympy as sp
import time
from IPython.display import display, HTML, Markdown, Image, YouTubeVideo
```

```python
x=sp.symbols("x",communtative = True)
```

```python
k=[]
d=0
a=0
b=0
c=0
st=time.time()

while(d<10):
    test=3.2
    while int(test)!=test:
        a=0
        while a==0:
            a=int(np.random.randint(-3,4,1))
        b=int(np.random.randint(-20,21,1))
        c=int(np.random.randint(-20,21,1))
        if b**2-4*a*c <0:
            continue
        else:
            
            pass
        test=np.sqrt(b**2-4*a*c)
    q=a*x**2+b*x+c
    answer=sp.solve(q,[x])
    display(Markdown(f"# 題目:\n## $$f(x)=0$$ \n## $$f(x)=({a})x^2+({b})x+({c})$$\n### x最大解為?\n(如果非整數用分數表示)"))
    qq=1
    while qq!=0:
        try:
            que1=input("x= ?\n")
            
            if "/" in que1:
                que1=que1.split("/")
                que1=int(que1[0])/int(que1[1])
            que1=float(que1)
            qq=0
        except:
            print("請輸入數字")
    if(float(que1)== float(max(answer))):
        print("答對了")
        k.append(1)
    else:
        print("答錯了")
        k.append(0)
    d=d+1
print("您一共答對了{0}，答錯了{1}題".format(sum(k),10-sum(k)))
ed=time.time()
print("一共花了{0}秒".format(round(ed-st),2))
```
