```python
import turtle as tu
from random import *
import time
```

```python
tu.showturtle()
tu.setup(800, 600, 0, 0)
tu.pensize(5)
tu.bgcolor("black") 
tu.pencolor('white')
tu.penup()
tu.goto(-300,200)
time.sleep(2)
tu.write("懷沁，",True,'Center',('Arial',24,'bold'))
time.sleep(2)
tu.goto(-150,150)
tu.write("雖然你不包養我，",True,'Center',('Arial',24,'bold'))
time.sleep(2)
tu.goto(-125,100)
tu.write("但還是恭喜你成年滿20歲了，",True,'Center',('Arial',24,'bold'))
time.sleep(2)
tu.goto(-150,50)
tu.write("祝你健康快樂，",True,'Center',('Arial',24,'bold'))
tu.goto(-150,0)
time.sleep(2)
tu.write("星期六記得投3號，",True,'Center',('Arial',24,'bold'))
tu.goto(-150,-50)
time.sleep(2)
tu.write("有需要的話，",True,'Center',('Arial',24,'bold'))
tu.goto(-125,-100)
time.sleep(2)
tu.write("妳外婆的證件我可以幫忙保管。",True,'Center',('Arial',24,'bold'))
tu.goto(0,-200)
time.sleep(2)
tu.write("呂修齊",True,'Center',('Arial',24,'bold'))
tu.speed(1)
tu.goto(35,-165)
tu.write("敬上",True,'Center',('Arial',12,'bold'))
time.sleep(5)
tu.clear()
time.sleep(1)
def ground():
    tu.hideturtle()
    
    for i in range(400):
        tu.pensize(randint(5,10))
        x = randint(-400,350)
        y = randint(-280,-1)
        r = -y/280
        g = -y/280
        b = -y/280
        tu.pencolor((r, g, b))
        tu.penup()
        tu.goto(x,y)
        tu.pendown()
        tu.forward(randint(40,100))

def snow():    
    tu.hideturtle()
    tu.pensize(2)
    #100朵花瓣
    for i in range(100):
        r = random()
        g = random()
        b = random()
        tu.pencolor(r, g, b)
        tu.penup()
        tu.setx(randint(-350,350))
        tu.sety(randint(1,270))
        tu.pendown()
        #dens表示花瓣数
        dens = randint(8,12)
        snowsize = randint(10,14)
        for j in range(dens):
            tu.forward(snowsize)
            tu.backward(snowsize)
            tu.right(360/dens)


def main():
    tu.setup(800, 600, 0, 0)
    tu.tracer(False)
    tu.bgcolor("black")    
    snow()
    ground()
    tu.tracer(True)


if __name__ == "__main__":
    main()

time.sleep(2)
tu.penup()
tu.pensize(5)
tu.sety(0)
tu.setx(0)
tu.color("white")
tu.write("懷沁生日快樂!!!",True,'Center',('Arial',48,'bold'))


tu.goto(250,-180)
tu.setheading(10)

def curvemove():
    for i in range(200):
        tu.right(1)
        tu.forward(1)
tu.color('maroon','darkmagenta')        
tu.speed(1000)
tu.penup()
tu.begin_fill()
tu.left(140)
tu.forward(111.65)
curvemove()
tu.left(120)
curvemove()
tu.forward(111.65)
tu.end_fill()
tu.done()
```
