# 第一題【猜數字遊戲】：
## 電腦由 1 到 100 「想」一個數字。你開始猜電腦想的數字, 太大或大小電腦會告訴你, 直到猜對。

```python
number=ran.randint(1,101)
aa=101
while(aa!=number):
    a= input("數字:")
    aa=float(a)
    if (aa > number):
        print("太大")
    elif(aa < number):
        print("太小")
    elif(aa==number):
        print("恭喜您，答對了!!")
```

# 第三題 【寫個函數讓字串倒過來排】

```python
def reverse(words):
    k=list(words)
    k.reverse()
    print(k)
```

# 挑戰題一【天干地支】：
## 年份是用天干地支，天干:甲乙丙丁戊己庚辛壬癸；地支:子醜寅卯辰巳午未申酉戌亥，2014年是甲午年，2015年是乙未年… 輸出2014~2073是什麼年

```python
天干=["甲","乙","丙","丁","戊","己","庚","辛","壬","癸"]
地支=["子","丑","寅","卯","辰","巳","午","未","申","酉","戌","亥",]
地支=地支[6:]+地支[0:6]
```

```python
a=天干*6
b=地支*5
for i in range(60):
    c=a[i]+b[i]+"====>"+"{}".format(2014+i)
    print(c)
```

# 寫個 tree 函數, 給一個數字就會畫出如下的畫面。

```python
def tree(n):
    for i in range(n):
        print(" "*(n-i)+"*"*(1+2*i)+" "*(n-i))
```

```python
tree(6)
```

```python
tree(16)
```
