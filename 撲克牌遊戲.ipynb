##### 遊戲方法: 首先先輸入A、B的名字，先從A開始隨機抽牌，A可以決定要不要繼續，如果差過21點B直接獲勝。接著換B的回合，一樣一次抽隨機抽一張牌，可決定要不要繼續。最後再來比較A與B的點數，點數大的獲勝。

```python
import time
```

```python
a=input("玩家A名字:")
b=input("玩家B名字:")
game=0
while game==0:
    time.sleep(1)
    print(f'{a}的回合開始')
    A=0
    aa=1
    while aa!=0:
        n=int(np.random.randint(1,14,1))
        A=int(A+n)
        time.sleep(1)
        if A>21:
            aa=0
            print(f'{a}的回合:\n抽到點數為{n}\n目前點數為{A}')
            time.sleep(1)
            print('你點數爆了')
            print(f'{b}獲勝')
            break
        print(f'{a}的回合:\n抽到點數為{n}\n目前點數為{A}')
        test=1
        while test==1:
            q=input("是否繼續(回答 是/否)")
            if q in ['是','否']:
                test=2
            else:
                print('請輸入正確指令')

        if q=='否':
            aa=0
    if A>21:
        break
    print(f"{a}的點數為{A}")
    time.sleep(1)
    print(f'{b}的回合開始')
    B=0
    bb=1
    while bb!=0:
        n=int(np.random.randint(1,14,1))
        B=int(B+n)
        time.sleep(1)
        if B>21:
            bb=0
            print(f'{a}的回合:\n抽到點數為{n}\n目前點數為{B}')
            time.sleep(1)
            print('你點數爆了')
            print(f'{a}獲勝')
            break
        print(f'{b}的回合:\n抽到點數為{n}\n目前點數為{B}')
        test=1
        while test==1:
            q=input("是否繼續(回答 是/否)")
            if q in ['是','否']:
                test=2
            else:
                print('請輸入正確指令')
        if q=='否':
            bb=0
    if B>21:
        break
    print(f"{b}的點數為{B}")
    time.sleep(1)
    print(f'{a}點數為{A} vs {b}點數為{B}')
    time.sleep(1)
    game=1
    if A>B:
        print(f'{a}獲勝')
    elif A<B:
        print(f'{b}獲勝')
    elif A==B:
        print("平手")
```
