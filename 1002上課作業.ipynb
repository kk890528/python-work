# 出加法10題，並計算秒數

```python
k=list()
a=0
st=time.time()
while(a<10):
    n1=int(np.random.randint(0,1000,1))
    n2=int(np.random.randint(0,1000,1))
    an=input("{0} + {2} = ".format(n1,k,n2))
    if (int(an)== n1+n2):
        print("答對了")
        k.append(1)
    else:
        print("答錯了")
        k.append(0)
    a=a+1
print("您一共答對了{0}，答錯了{1}題".format(sum(k),10-sum(k)))
ed=time.time()
print("一共花了{0}秒".format(round(ed-st),2))
```

# 計算從我出生到現在，已經過了多久

```python
my_birth=time.strptime('2000-04-28 10:00', '%Y-%m-%d %H:%M')
```

```python
now=time.time()
interval=now-time.mktime(my_birth)
```

```python
def birth():
    a=0
    while a!=1: 
        now=time.time()
        interval=now-time.mktime(my_birth)
        m, s = divmod(interval, 60)
        h, m = divmod(m, 60)
        d, h = divmod(h, 24)
        y, d = divmod(d, 365)
        k="{0}年{1}天{2}小時{3}分{4}秒".format(int(y),int(d),int(h),int(m),int(s))
        print(k)
        time.sleep(1)
        
```
