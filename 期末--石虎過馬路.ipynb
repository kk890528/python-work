# 小遊戲:石虎過馬路
### 遊戲介紹:在苗栗，石虎過馬路都危機重重，容易被路殺，請幫助石虎過馬路
#### 畫面上會有車子，每台車子的速度不定(5-14格)，每次可以向前移動-2到2格，每移動一次便會生成更多車子，當石虎移動10格，變成功通過馬路，如果到達的剛好有車子經過，便會被路殺。

```python
x=0
y=0
plt.scatter(x,y,c="r")
plt.annotate("石虎",(x,y))
plt.xlim(-20,20)
plt.ylim(-0.5,10.5)
xm,ym=np.meshgrid(np.arange(-20,21,8),np.arange(1,10,2))
P=np.c_[xm.ravel(),ym.ravel()]
plt.scatter(P[:,0],P[:,1],marker="<",c="blue",alpha=0.3)
xm,ym=np.meshgrid(np.arange(-19,22,8),np.arange(1,10,2))
P=np.c_[xm.ravel(),ym.ravel()]
plt.scatter(P[:,0],P[:,1],marker="<",c="blue",alpha=0.3)

xm,ym=np.meshgrid(np.arange(-23,20,8),np.arange(2,10,2))
P=np.c_[xm.ravel(),ym.ravel()]
plt.scatter(P[:,0],P[:,1],marker=">",c="green",alpha=0.3)
xm,ym=np.meshgrid(np.arange(-22,20,8),np.arange(2,10,2))
P=np.c_[xm.ravel(),ym.ravel()]
plt.scatter(P[:,0],P[:,1],marker=">",c="green",alpha=0.3)

x1,x2,x3=np.random.choice(range(0,15),3)
y1,y2,y3=np.random.choice(range(1,10),3)
X=[x1,x2,x3]
Y=[y1,y2,y3]
Y=np.array(Y)
X=np.array(X)
try:
    X[Y %2==0]=abs(X[ Y %2==0])*-1
except IndexError:
    pass
plt.scatter(x1,y1,c="black")
plt.annotate("車",(x1,y1))
plt.scatter(x2,y2,c="black")
plt.annotate("車",(x2,y2))
plt.scatter(x3,y3,c="black")
plt.annotate("車",(x3,y3))
speed=np.random.choice(range(5,16),3)
plt.show()
qq=0
while y<10 and qq !=10:
    k=input("你要如何移動?(可輸入-2到2(包括0)     ")
    try:
        k=int(k)
    except:
        print("請輸入數字")
        continue
    if k not in [1,2,0,-1,-2]:
        print("請輸入正確數字")
        continue
    if y+k<0  :
        print("不能再退了")
        continue
    else:
        y=y+k
        try:
            X[Y%2==0]=X[Y%2==0]+speed[Y%2==0]
        except IndexError:
            pass
        try:
            X[Y%2==1]=X[Y%2==1]-speed[Y%2==1]
        except IndexError:
            pass
        n=np.random.randint(1,7)
        
        x_x=np.random.choice(range(15,21),n)
        y_y=np.random.choice(range(1,10),n)
        try:                                                  
            x_x[y_y%2==0]=x_x[y_y%2==0]*-1
        except:
            pass
        for i in range(n):
            speed=np.append(speed,np.random.choice(range(5,15)))
        X=np.append(X,x_x)
        Y=np.append(Y,y_y)
    if y in Y and y<10:
        if y%2==1 and min(X[Y==y]*(X[Y==y]+speed[Y==y]))<=0:
            a=0
        elif y%2==0 and min(X[Y==y]*(X[Y==y]-speed[Y==y]))<=0:
            a=0
        else:
            a=1
        if a!=1:
            print("您被路殺了")
            qq=10
            plt.cla()
            plt.scatter(x,y,c="r",marker="x")
            plt.annotate("石虎屍體",(x,y))
            plt.xlim(-20,20)
            plt.ylim(-0.5,10.5)
            plt.scatter(X,Y,c='black')
            
            xm,ym=np.meshgrid(np.arange(-20,21,8),np.arange(1,10,2))
            P=np.c_[xm.ravel(),ym.ravel()]
            plt.scatter(P[:,0],P[:,1],marker="<",c="blue",alpha=0.3)
            xm,ym=np.meshgrid(np.arange(-19,22,8),np.arange(1,10,2))
            P=np.c_[xm.ravel(),ym.ravel()]
            plt.scatter(P[:,0],P[:,1],marker="<",c="blue",alpha=0.3)

            xm,ym=np.meshgrid(np.arange(-23,20,8),np.arange(2,10,2))
            P=np.c_[xm.ravel(),ym.ravel()]
            plt.scatter(P[:,0],P[:,1],marker=">",c="green",alpha=0.3)
            xm,ym=np.meshgrid(np.arange(-22,20,8),np.arange(2,10,2))
            P=np.c_[xm.ravel(),ym.ravel()]
            plt.scatter(P[:,0],P[:,1],marker=">",c="green",alpha=0.3)
            
            for i in range(len(X)):
                plt.annotate("車",(X[i],Y[i]))
            plt.show()
        else:
            plt.cla()
            plt.scatter(x,y,c="red")
            plt.annotate("石虎",(x,y))
            plt.scatter(X,Y,c="black")
            plt.xlim(-20,20)
            plt.ylim(-0.5,10.5)
            
            xm,ym=np.meshgrid(np.arange(-20,21,8),np.arange(1,10,2))
            P=np.c_[xm.ravel(),ym.ravel()]
            plt.scatter(P[:,0],P[:,1],marker="<",c="blue",alpha=0.3)
            xm,ym=np.meshgrid(np.arange(-19,22,8),np.arange(1,10,2))
            P=np.c_[xm.ravel(),ym.ravel()]
            plt.scatter(P[:,0],P[:,1],marker="<",c="blue",alpha=0.3)

            xm,ym=np.meshgrid(np.arange(-23,20,8),np.arange(2,10,2))
            P=np.c_[xm.ravel(),ym.ravel()]
            plt.scatter(P[:,0],P[:,1],marker=">",c="green",alpha=0.3)
            xm,ym=np.meshgrid(np.arange(-22,20,8),np.arange(2,10,2))
            P=np.c_[xm.ravel(),ym.ravel()]
            plt.scatter(P[:,0],P[:,1],marker=">",c="green",alpha=0.3)
            
            for i in range(len(X)):
                plt.annotate("車",(X[i],Y[i]))
            plt.show()
    else:
        plt.cla()
        plt.scatter(x,y,c="r")
        plt.annotate("石虎",(x,y))
        plt.scatter(X,Y,c="black")
        plt.xlim(-20,20)
        plt.ylim(-0.5,10.5)
        xm,ym=np.meshgrid(np.arange(-20,21,8),np.arange(1,10,2))
        P=np.c_[xm.ravel(),ym.ravel()]
        plt.scatter(P[:,0],P[:,1],marker="<",c="blue",alpha=0.3)
        xm,ym=np.meshgrid(np.arange(-19,22,8),np.arange(1,10,2))
        P=np.c_[xm.ravel(),ym.ravel()]
        plt.scatter(P[:,0],P[:,1],marker="<",c="blue",alpha=0.3)

        xm,ym=np.meshgrid(np.arange(-23,20,8),np.arange(2,10,2))
        P=np.c_[xm.ravel(),ym.ravel()]
        plt.scatter(P[:,0],P[:,1],marker=">",c="green",alpha=0.3)
        xm,ym=np.meshgrid(np.arange(-22,20,8),np.arange(2,10,2))
        P=np.c_[xm.ravel(),ym.ravel()]
        plt.scatter(P[:,0],P[:,1],marker=">",c="green",alpha=0.3)
        for i in range(len(X)):
            plt.annotate("車",(X[i],Y[i]))
        plt.show()
    if y>=10:
        print("恭喜您，成功通過一次馬路")
```
