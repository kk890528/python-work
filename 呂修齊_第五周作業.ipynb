第一題_算閏年：
請定義一個函式，讀進一個年份，輸出是不是閏年。

提示：
目前使用的格里高利曆閏年規則如下：

4的倍數是可能的。
100的倍數是不可能的。
400的倍數是可能的。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def year():
    y=input("今年西元幾年?     ")
    y=int(y)
    if divmod(y,4)[1] !=0:
        print("這不是閏年")
    elif divmod(y,400)[1] ==0:
        print("這是閏年")
    elif divmod(y,100)[1] ==0:
        print("這不是閏年")
    else:
        print("這是閏年")
```

```python
year()
```

挑戰題一_Bubble Sort
Geoffrey是個喜歡自動化的老師，某次期中考過後，他想要把學生的成績用電腦自動排序，於是他請Andrew幫忙開發這個高難度的程式…

限制：不可以使用Python內建的sort()

```python
grades=list(np.random.randint(0,101,9))
sg=[]

for i in range(len(grades)):
    for j in range(len(grades)):
        if grades[j]==min(grades):
            sg.append(grades[j])
            grades.remove(grades[j])
            grades.append(101)
        if len(sg)>=9:
            print("已排序")
            break
    if len(sg)>=9:
            break
```
