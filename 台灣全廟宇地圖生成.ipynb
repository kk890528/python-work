# 利用政府開放資料，將臺灣全寺廟的xml檔變成DataFrame格式後，依照各種不同的神，畫出全台灣廟宇分布圖

```python
import xml.etree.cElementTree as ET
```

```python
from urllib.request import urlopen
```

```python

doc =ET.parse('temple.xml')
```

```python
root=doc.getroot()
```

```python
data=root.findall('OpenData_3')
```

```python
cols=[]
for elem in data[1].iter():
    cols.append(elem.tag)
```

```python
cols.pop(0)
```

```python
temples=pd.DataFrame(columns=cols)
```

```python
for i in range(len(data)):
    temple=[]
    for j in range(len(cols)):
        try:
            information=data[i].find(cols[j]).text
        except:
            information='無資料'
        temple.append(information)
    temples.loc[i]=temple
```

```python
temples.head()
```

```python
religions=list(temples['主祀神祇'].value_counts().index)
```

```python
religions
```

```python
temples.rename(columns={'WGS84Y':'經度','WGS84X':'緯度'},inplace=True)
```

```python
import folium
from folium import Marker
from folium import IFrame
```

```python
def utf2asc(s):
    return str(str(s).encode('ascii', 'xmlcharrefreplace'))[2:-1]
```

```python
heading3 = """<h3>{}</h3>""".format
iframe = IFrame(html=heading3(utf2asc('寺廟')),width=100,height=50)
popup = folium.Popup(iframe)
```

```python
iframe
```

```python
def god_map(religion):
    m1= folium.Map(location=[23.7,120.5], tiles='OpenStreetMap', zoom_start=7.5)
    god_select=temples[temples['主祀神祇']==religion]
    for idx, row in god_select.iterrows():
        tem=row['寺廟名稱']
        heading3 = """<h3>{}</h3>""".format
        iframe = IFrame(html=heading3(utf2asc(tem)),width=100,height=50)
        popup = folium.Popup(iframe)
        try:
            loca=[row['經度'],row['緯度']]
            folium.Marker(location=loca,
            popup=popup
                         ).add_to(m1)
        except:
            pass
    return m1
```

```python
from ipywidgets import interact
```

```python
interact(god_map,religion=religions)
```
