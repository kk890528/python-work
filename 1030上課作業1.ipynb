# 篩出男性與女性對欲在公園或海灘對於陌生人全裸的看法

```python
male=df[df["請問您的性別認同為?"]=="男性"]
female=df[df["請問您的性別認同為?"]=="女性"]
```

```python
sex=pd.DataFrame()
```

```python
sex["男性"]=male["請問您是否可以接受在公園或海灘看到有陌生人全裸?"].value_counts()
sex["女性"]=female["請問您是否可以接受在公園或海灘看到有陌生人全裸?"].value_counts()
```

```python
sex["男性(%)"]=round(sex["男性"]/sex["男性"].sum()*100,2)
sex["女性(%)"]=round(sex["女性"]/sex["女性"].sum()*100,2)
```
