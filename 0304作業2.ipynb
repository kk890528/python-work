# 利用文化部的資料，畫出從有資料以來(2018年8月)的票房累積競賽圖，新增利用Datetime增加日期變化顯示

```python
def number(n):
    l=[]
    for i in range(len(n)):
        try:
            l.append(int(n[i].replace(',','')))
        except:
            l.append(n[i])
    return l
```

```python
def word(w):
    l=[]
    for i in range(len(w)):
        try:
            l.append(w[i].replace(' ',''))
        except:
            l.append(w[i])
    return l
```

```python
all_data=[]
for i in range(len(allFileList)):
    all_data.append(pd.read_excel(f'C:/Users/Michael/Desktop/python/python下學期/movie/{allFileList[i]}'))
```

```python
for i in range(len(all_data)):
    all_data[i]['累計銷售金額']=number(all_data[i].iloc[:,-1])
```

```python
for i in range(len(all_data)):
    all_data[i]['中文片名']=word(all_data[i].iloc[:,2])
```

```python
all_movie=[]
for i in range(len(all_data)):
    for s in range(len(all_data[i])):
        all_movie.append(all_data[i]['中文片名'][s])
```

```python
all_movie=pd.unique(all_movie)
len(all_movie)
```

```python
def randomcolor(n):
    colorArr = ['1','2','3','4','5','6','7','8','9','A','B','C','D','E','F']
    colors=[]
    for s in range(n):
        col = ""
        for i in range(6):
            col += colorArr[np.random.randint(0,15)]
            color="#"+col
        colors.append(color)
    return colors
```

```python
movie_colors=randomcolor(len(all_movie))
```

```python
cols = dict(zip(
    all_movie,
    movie_colors
))
```

```python
k=pd.DataFrame(all_data[(0)].groupby('中文片名').累計銷售金額.max()).reset_index()
kk=k.sort_values(by='累計銷售金額',ascending=True).tail(10)
movies=[kk]
for i in range(len(all_data)-1):
    m1=movies[i]
    m2=all_data[(i+1)]
    m=pd.concat([m1,m2])
    mm=pd.DataFrame(m.groupby('中文片名').累計銷售金額.max()).reset_index()
    mm=mm.sort_values(by='累計銷售金額',ascending=True).tail(10)
    movies.append(mm)
```

```python
t1 = datetime.datetime(2018, 8, 6) 
t2 = datetime.datetime(2018, 8, 12)
td = datetime.timedelta(days=7)
```

```python
times=[]
for i in range(len(all_data)):
    tt1=t1.strftime('%Y-%m-%d')
    tt2=t2.strftime('%Y-%m-%d')
    times.append(f"{tt1}\n     to\n{tt2}")
    t1=t1+td
    t2=t2+td
```

```python
for i in range(len(movies)):
    movies[i]['時間']=times[i]
    
```

```python
movies2=pd.concat(movies)
movies2=movies2.reset_index()
fr=list(movies2['時間'].unique())
```

```python
def time_barh(movie):
    colors=[]
    mo=movies2[movies2['時間']==movie]
    mo=mo.reset_index()
    plt.cla()
    for x in mo['中文片名']:
        colors.append(cols[x])
    plt.barh(data=mo,width='累計銷售金額',y='中文片名',height=0.8,color=colors)
    plt.grid(which='major', axis='x', linestyle='-')
    plt.title('2018年8月開始國片票房累積變化圖')
    plt.xlabel('票房(千萬)')
    plt.text(mo['累計銷售金額'][len(mo)-1],0.2,movie,size=36,weight=1000,ha='right')
    plt.box(False)
```

```python
import matplotlib as mpl
mpl.rcParams['figure.max_open_warning'] = 0
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
animator = animation.FuncAnimation(fig, time_barh, frames=fr)
animator.save("movie.mp4", fps=2)
```
