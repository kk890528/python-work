### 寫一個小程式，把大於30天沒碰的資料丟進回收桶

```python
len(os.listdir('C:\\Users\Michael\Downloads'))  #全檔案
```

```python
d=os.listdir('C:\\Users\Michael\Downloads')[0]
```

```python
os.path.getctime(f'C:\\Users\Michael\Downloads\{d}')
```

```python
for r in os.listdir('C:\\Users\Michael\Downloads'):
    t= os.path.getctime(f'C:\\Users\Michael\Downloads\{r}')
    t= int(t)
    nt= int(time.time())
    an= nt-t
    if an > 2592000: # 清除30天前的檔案 (2592000秒)
        send2trash.send2trash(f'C:\\Users\Michael\Downloads\{r}')
```
