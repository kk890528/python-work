# 遊戲:石虎過馬路
## 苗栗縣某議員說肇因於石虎太多，所以常常被車撞，請幫助石虎們過馬路吧
### 這星期新增雙向道(奇數的話向左行，偶數向右)

```python
x=0
y=0
plt.scatter(x,y,c="r")
plt.annotate("leopard cat",(x,y))
plt.xlim(-20,20)
plt.ylim(-0.5,10.5)
x1,x2,x3=np.random.choice(range(0,15),3)
y1,y2,y3=np.random.choice(range(1,10),3)
X=[x1,x2,x3]
Y=[y1,y2,y3]
Y=np.array(Y)
X=np.array(X)
try:
    X[Y %2==0]=abs(X[ Y %2==0])*-1
except IndexError:
    pass
plt.scatter(x1,y1,c="black")
plt.annotate("car",(x1,y1))
plt.scatter(x2,y2,c="black")
plt.annotate("car",(x2,y2))
plt.scatter(x3,y3,c="black")
plt.annotate("car",(x3,y3))
speed=np.random.choice(range(5,16),3)
plt.show()
qq=0
while y<10 and qq !=10:
    k=input("你要如何移動?(可輸入-2到2(包括0)     ")
    k=int(k)
    if y+k<0:
        print("不能再退了")
        continue
    else:
        y=y+k
        try:
            X[Y%2==0]=X[Y%2==0]+speed[Y%2==0]
        except IndexError:
            pass
        try:
            X[Y%2==1]=X[Y%2==1]-speed[Y%2==1]
        except IndexError:
            pass
        n=np.random.randint(1,5)
        X=np.append(X,(np.random.choice(range(15,21),n)))
        Y=np.append(Y,(np.random.choice(range(1,10),n)))
        try:
            X[Y %2==0]=list(abs(X[ Y %2==0])*-1)
        except IndexError:
            pass

    for i in range(n):
        speed=np.append(speed,np.random.choice(range(5,15)))
    if y in Y:
        if y%2==1:
            a=min(X[Y==y]*(X[Y==y]+speed[Y==y]))<0
        elif y%2==0:
            a=min(X[Y==y]*(X[Y==y]-speed[Y==y]))<0
        else:
            a=1
        if a!=1:
            print("您被路殺了")
            qq=10
            plt.cla()
            plt.scatter(x,y,c="r",marker="x")
            plt.annotate("leopard cat",(x,y))
            plt.xlim(-20,20)
            plt.ylim(-0.5,10.5)
            plt.scatter(X,Y,c="black")
            for i in range(len(X)):
                plt.annotate("car",(X[i],Y[i]))
            plt.show()
        else:
            plt.cla()
            plt.scatter(x,y,c="r")
            plt.annotate("leopard cat",(x,y))
            plt.scatter(X,Y,c="black")
            plt.xlim(-20,20)
            plt.ylim(-0.5,10.5)
            for i in range(len(X)):
                plt.annotate("car",(X[i],Y[i]))
            plt.show()
    else:
        plt.cla()
        plt.scatter(x,y,c="r")
        plt.annotate("leopard cat",(x,y))
        plt.scatter(X,Y,c="black")
        plt.xlim(-20,20)
        plt.ylim(-0.5,10.5)
        for i in range(len(X)):
            plt.annotate("car",(X[i],Y[i]))
        plt.show()
    if y>=10:
        print("恭喜您，成功通過一次馬路")
```
