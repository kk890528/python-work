# map and filter

```python
items=[500,30,2000]
c=30.8594
```

```python
egg=[]
for i in range(len(items)):
    k=items[i]*c
    egg.append(k)
```

```python
egg
```

```python
def exchange(p):
    return p*c
```

```python
exchange(500)
```

```python
egg=list(map(exchange,items))
```

```python
egg
```

# lambda 救急函數

```python
list(map(lambda p:p*c,items))
```

# list comprehesion

```python
[p*c for p in items]
```

# filter

```python
grades=[50,59,65,80,100,34,40,88]
```

```python
grades[1]>60
```

```python
def spam(g):
    return g >=60
```

```python
spam(70)
```

```python
list(filter(spam,grades))
```

```python
list(filter(lambda g:g>=60,grades))
```

```python
grades=np.array(grades)
grades[grades>60]
```

# 更完整print控制

```python
egg="(0_0)"
spam=f"xx{egg}xx"
print(spam)
```

## round 1

```python
spam=f"xx{egg:20s}xx"
print(spam)
```

## round 2

```python
spam=f"xx{egg:>20s}xx"
print(spam)
```

## round 3

```python
spam=f"xx{egg:^20s}xx"
print(spam)
```

```python
level=5
for n in range(level):
    s="*"*(2*n+1)
    print(f"{s:^21s}")
```

```python
spam=f"xx{egg:*^20s}xx"
print(spam)
```

```python
n=20
spam=f"{str(n):0>3s}"
print(spam)
```

```python
n=87
spam="000"+str(n)
print(spam[2:])
```

```python
n=str(87)
spam=n.zfill(3)
print(spam)
```
