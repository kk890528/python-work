# 中國信用制度
## 為了成為1984的老大哥，中國將要開始實施信用制度，初始點數為1000

```python
class credit:
    def __init__(self, name):
        self.name = name
        self.credit_value = 1000
        self.record=pd.DataFrame(columns=["原因","增減"],)
    def deposit(self,reason,amount):  #增加信用用動作: amount代表存入金額
        if amount <= 0:
            raise ValueError('')
        self.record.loc[self.record.shape[0]+1]=[reason,f"+{amount}"]
        self.credit_value += amount
        
    def withdraw(self,reason,amount): #扣信用動作: amount代表取款金額
        self.record.loc[self.record.shape[0]+1]=[reason,f"-{amount}"]
        if amount <= self.credit_value:
            self.credit_value -= amount
        else:
            self.credit_value=0
            print("親~你信用破產了")
    def __repr__(self):
        value=f"姓名:  {self.name}\n\n目前您的信用點數為:  {self.credit_value}點"
        return(value)
```
