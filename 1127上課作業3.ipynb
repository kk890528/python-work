# 使用中研院2016台灣社會變遷資料庫公民國家組，劃出不同自我族群認同所支持的總統候選人

[資料網址](https://srda.sinica.edu.tw/datasearch_detail.php?id=1065)

```python
df, meta = pyreadstat.read_sav('tscs2016q2.sav')
```

```python
a=df[df["a9"]==1]#閩南
b=df[df["a9"]==2]#客家
c=df[df["a9"]==3]#原住民
d=df[df["a9"]==4]#外省人
e=df[df["a9"]==5]#其他
```

# H5.請問您投票給哪一組候選人？(提示卡 29) □(01)朱立倫與王如玄     □(02)蔡英文與陳建仁     □(03)宋楚瑜與徐欣瑩   □(04)投廢票

```python
seaborn.barplot（x=np.unique(a['h8']),y=a['h8']）
```

```python
a1=pd.value_counts(a['h5'])[np.unique(a['h5'])]
b1=pd.value_counts(b['h5'])[np.unique(b['h5'])]
c1=pd.value_counts(c['h5'])[np.unique(c['h5'])]
d1=pd.value_counts(d['h5'])[np.unique(d['h5'])]
e1=pd.value_counts(e['h5'])[np.unique(e['h5'])]
```

```python
target_names={1:'朱立倫與王如玄' ,2:'蔡英文與陳建仁' ,3:'宋楚瑜與徐欣瑩' ,4:'投廢票',96:'跳答',97:'忘記了'
,98:'拒答'}
```

```python
f, ax= plt.subplots(figsize = (10, 10))
ax.set_title('閩南人2016總統大選投票')
sns.barplot(x=a1.index,y=a1.values,hue=a1.index.map(target_names))
```

```python
f, ax= plt.subplots(figsize = (10, 10))
ax.set_title('客家人2016總統大選投票')
sns.barplot(x=b1.index,y=b1.values,hue=a1.index.map(target_names))
```

```python
f, ax= plt.subplots(figsize = (10, 10))
ax.set_title('原住民2016總統大選投票')
sns.barplot(x=c1.index,y=c1.values,hue=c1.index.map(target_names))
```

```python
f, ax= plt.subplots(figsize = (10, 10))
ax.set_title('外省人2016總統大選投票')
sns.barplot(x=d1.index,y=d1.values,hue=d1.index.map(target_names))
```

```python
f, ax= plt.subplots(figsize = (10, 10))
ax.set_title('其他種人2016總統大選投票')
sns.barplot(x=e1.index,y=e1.values,hue=e1.index.map(target_names))
```
