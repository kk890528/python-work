# 打入一串字串，可以幫你把句子變成晶晶體，讓你去商會演講不再害怕

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from googletrans import Translator
import math
```

```python
def gingin(word):
    l_word=list(word)
    n=np.random.randint(1,len(l_word),math.ceil(len(l_word)/5))
    for i in n:
        translator = Translator()
        tran=translator.translate(l_word[i],src='zh-TW' ,dest = "en")
        l_word[i]=tran.text
    print(''.join(l_word))
```

```python
gingin('二二八跟八二三是不一樣的')
```
