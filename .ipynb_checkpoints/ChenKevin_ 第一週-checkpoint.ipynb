# 挑戰題一：買相機

```python
x=[100,150,230,80]
x_add=[x[0], x[0]+x[1], x[0]+x[1]+x[2], x[0]+x[1]+x[2]+x[3]]
plot([1,2,3,4], x_add)
```

```python
pie(x, labels=["equip1", "equip2", "equip3", "equip4"], autopct="%1.1f%%")
```

# 挑戰題二：自己出！

## 輸入英呎英吋, 換算成公制並印出

```python
feet = input("請輸入英呎:")
inch = input("請輸入英吋:")
print(str(feet) + "英呎" + str(inch) + "英吋 = " + str((float(feet)*12+float(inch))*2.54) + "公分")
```
