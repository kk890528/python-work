# 挑戰題一：買相機
## 財團法人大兔兔文教基金會要添購攝影設備。他們預計在未來四個月內分別添購四項價格為 100k,150k,230k,80k的設備，請以折線圖畫出累積消費額，並以圓餅圖畫出各器材所佔比例

```python
x=[1,2,3,4]
equip1=100
equip2=150
equip3=230
equip4=80
y=[equip1,equip1+equip2,equip1+equip2+equip3,equip1+equip2+equip3+equip4]
plt.plot(x,y)
```

```python
x=[equip1,equip2,equip3,equip4]
plt.pie(x,labels=["a","b","c","d"],autopct = "%1.1f%%")
```
