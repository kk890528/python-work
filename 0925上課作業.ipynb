# 高雄市議會質詢，因應市長要準時下班的責任制，所以最多只能有12位議員可以質詢，設一個函數，n=參與抽籤的議員(最多66，扣掉正副議長64)，來看預估參加幾次抽到的機率才會大於99%

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from ipywidgets import interact
```

```python
def p(g):
    pp=((g-12)/g)
    return(pp)
```

```python
def question(n):
    k=list()
    c=1
    time=p(n)
    while (time>0.01):
        time=time**2
        c=c+1
    print("要參加"+str(c)+"次質詢")
```

```python
interact(question,n=(1,64))
```
