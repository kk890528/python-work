# 拍拍機器人

```python
k="kk"
while k!="":
    k=input(">>")
    if("開心"in k or"快樂"in k or"幸福"in k or"高興"in k or"幸運"in k or"帥"in k or"美"in k or"俊" in k):
        print("拍拍機器人:乾我屁事喔!")
        print("  ")
        print("  ")
        print("---"*10)
        print('想結束就打"bye" ')
    elif( k==""):
        print("拍拍機器人:跩什麼跩!不理你了")
    elif("愛"in k or"喜歡"in k or"欣賞"in k):
        print("拍拍機器人:阿餒母湯喔")
        print("  ")
        print("  ")
        print("---"*10)
        print('想結束就打"bye" ')
    elif(k == "bye"):
        break
    else:
        print("拍拍機器人:拍拍")
        print("  ")
        print("  ")
        print("---"*10)
        print('想結束就打"bye" ')
        
```
