# 出加法、乘法或剪法10題，並計算秒數

```python
k=[]
a=0
b=0
c=0
st=time.time()
while(a<10):
    ch=np.random.randint(0,3,1)
    n1=int(np.random.randint(0,1000,1))
    n2=int(np.random.randint(0,1000,1))
    if ch==0:
        an=input("{} + {} = ".format(n1,n2))
        if(int (an)== n1+n2):
            print("答對了")
            k.append(1)
        else:
            print("答錯了")
            k.append(0)
        
    if ch==1:
        an=input("{} - {} = ".format(n1,n2))
        if (int(an)== n1-n2):
            print("答對了")
            k.append(1)
        else:
            print("答錯了")
            k.append(0)
    
    if ch==2:
        an=input("{} x {} = ".format(n1,n2))
        
        if(int(an)== n1*n2):
            print("答對了")
            k.append(1)
        else:
            print("答錯了")
            k.append(0)
    a=a+1
print("您一共答對了{0}，答錯了{1}題".format(sum(k),10-sum(k)))
ed=time.time()
print("一共花了{0}秒".format(round(ed-st),2))
```

# 馬路如虎口，苗栗縣某議員說石虎太多才會被車撞，請幫助石虎過馬路。
## 一格等於10m 移動一次等於一秒

```python
x=0
y=0
plt.scatter(x,y,c="r")
plt.annotate("leopard cat",(x,y))
plt.xlim(-20,20)
plt.ylim(-0.5,10.5)
x1,x2,x3=np.random.choice(range(0,15),3)
y1,y2,y3=np.random.choice(range(1,10),3)
plt.scatter(x1,y1,c="black")
plt.annotate(f"car({50}m/s)",(x1,y1))
plt.scatter(x2,y2,c="black")
plt.annotate(f"car({50}m/s)",(x2,y2))
plt.scatter(x3,y3,c="black")
plt.annotate(f"car({50}m/s)",(x3,y3))
X=[x1,x2,x3]
Y=[y1,y2,y3]
plt.show()
qq=0
while y<10 and qq !=10:
    k=input("你要如何移動?(可輸入-2到2(包括0)     ")
    k=int(k)
    if y+k<0:
        print("不能再退了")
        continue
    else:
        y=y+k
    X=list(np.array(X)-5)
    n=np.random.randint(1,3)
    X=X+list(np.random.choice(range(15,21),n))
    Y=Y+list(np.random.choice(range(1,10),n))
    if y in Y:
        if -4 in np.array(X)[y==np.array(Y)]or-3 in np.array(X)[y==np.array(Y)]or -1 in np.array(X)[y==np.array(Y)] or -2 in np.array(X)[y==np.array(Y)] or 0 in np.array(X)[y==np.array(Y)]:
            print("您被路殺了")
            qq=10
            plt.cla()
            plt.scatter(x,y,c="r",marker="x")
            plt.xlim(-20,20)
            plt.ylim(-0.5,10.5)
            plt.scatter(X,Y,c="black")
            for i in range(len(X)):
                plt.annotate(f"car({50}m/s)",(X[i],Y[i]))
            plt.show()
        else:
            plt.cla()
            plt.scatter(x,y,c="r")
            plt.scatter(X,Y,c="black")
            plt.xlim(-20,20)
            plt.ylim(-0.5,10.5)
            for i in range(len(X)):
                plt.annotate(f"car({50}m/s)",(X[i],Y[i]))
            plt.show()
    else:
        plt.cla()
        plt.scatter(x,y,c="r")
        plt.scatter(X,Y,c="black")
        plt.xlim(-20,20)
        plt.ylim(-0.5,10.5)
        for i in range(len(X)):
            plt.annotate(f"car({50}m/s)",(X[i],Y[i]))
        plt.show()
    if y>=10:
        print("恭喜您，成功通過一次馬路")
```
